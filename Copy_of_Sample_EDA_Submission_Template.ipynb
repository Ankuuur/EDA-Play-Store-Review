To make a Size a Numeric Column

Observation:

1.Varies with device

2.K

3.M

````python
# Write your code to make your dataset analysis ready.
playstore_df['Size'].isnull().sum()
````

Verify the number of values and 3 different kind of values

````python
# Find the values of M having in them
playstore_df['Size'].loc[playstore_df['Size'].str.contains('M')].value_counts().sum()
````

````python
# Find the values of k having in them
playstore_df['Size'].loc[playstore_df['Size'].str.contains('k')].value_counts().sum()
````

````python
# Find the values of Varies with devices having in them
playstore_df['Size'].loc[playstore_df['Size'].str.contains('Varies with device')].value_counts().sum()
````

````python
#Convert the whole Size column to Bytes
# Lets define a function
def convert_size(size):
    if isinstance(size,str):
        if 'k' in size:
            return float(size.replace('k',"")) * 1024
        elif 'M' in size:
            return float(size.replace('M',"")) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
````

````python
# lets apply this function
playstore_df['Size'] = playstore_df['Size'].apply(convert_size)
````
